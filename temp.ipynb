```python
from torchvision.transforms import transforms
from torchvision.datasets import ImageFolder

from pathlib import Path

from torch.utils.data import Dataset, DataLoader


TRAIN_DATASET_DIR = Path("/home/jongha/projects/FGVC-RA-CNN/data/Iron-Scraps/my_classification/vanilla/split_dataset_25pct/train")
VAL_DATASET_DIR = Path("/home/jongha/projects/FGVC-RA-CNN/data/Iron-Scraps/my_classification/vanilla/split_dataset_25pct/val")


def get_dataset(train_dataset_dir: Path, val_dataset_dir: Path) -> tuple[Dataset, Dataset]:    
    train_transform = transforms.Compose([
        transforms.Resize(size=(224, 224)),
        transforms.RandomHorizontalFlip(p=0.5),
        transforms.ColorJitter(
            brightness=0.2, contrast=0.2, saturation=0.2, hue=0.2
        ),
        transforms.RandomVerticalFlip(p=0.5),
        transforms.RandomAffine(degrees=15, scale=(0.9, 1.1)),
        transforms.ToTensor(),
        transforms.Normalize(
            mean = [0.485, 0.456, 0.406],
            std = [0.229, 0.224, 0.225]),
    ])
    val_transform = transforms.Compose([
        transforms.Resize(size=(224, 224)),
        transforms.ToTensor(),
        transforms.Normalize(
            mean = [0.485, 0.456, 0.406],
            std = [0.229, 0.224, 0.225]),
    ])    

    train_dataset = ImageFolder(
        root=train_dataset_dir,
        transform=train_transform,
    )
    val_dataset = ImageFolder(
        root=val_dataset_dir,
        transform=val_transform,
    )

    return train_dataset, val_dataset
```

```python
train_dataset, val_dataset = get_dataset(
    train_dataset_dir=TRAIN_DATASET_DIR, 
    val_dataset_dir=VAL_DATASET_DIR,
)
```

```python
import yaml


HYPERPARAM_PATH = Path("hyper-parameters.yaml")


def load_hyperparams(hyperparams: Path):
    with open(file="hyper-parameters.yaml", mode="r", encoding="utf-8") as f:
        hyperparams = yaml.safe_load(stream=f)

    return hyperparams


def get_dataloader(hyperparam_path: Path, train_dataset: Dataset, val_dataset: Dataset) -> tuple[DataLoader, DataLoader]:
    hyperparams = load_hyperparams(hyperparams=hyperparam_path)
    batch_size = hyperparams["training"]["BATCH_SIZE"]

    train_loader = DataLoader(
        dataset=train_dataset,
        batch_size=batch_size,
        num_workers=4,
        pin_memory=True,
        shuffle=True,
    )
    val_loader = DataLoader(
        dataset=val_dataset,
        batch_size=batch_size,
        num_workers=4,
    )

    return train_loader, val_loader
```

```python
train_loader, val_loader = get_dataloader(
    hyperparam_path=HYPERPARAM_PATH,
    train_dataset=train_dataset,
    val_dataset=val_dataset,
)
```
